```python
df_labels = get_df_filenames(output_folder)

print(df_labels.sample(5))
```

```python
path_data = '../data/FairFace/processed_images'
```

```python
df_labels_with_faces = capture_faces_fan(path_data, df_labels)
```

```python
df_labels_with_faces_extended = capture_faces_mtcnn(path_data, df_labels)
```

```python
df_labels_recognized = pd.merge(
    df_labels_with_faces_extended,
    df_labels_with_faces,
    on=['filename'],
    how='inner',
    suffixes=('_extended', '_original')
)
```

```python
df_labels_recognized_cleaned = df_labels_recognized[(df_labels_recognized['face_found_mtcnn'] != 0) & (df_labels_recognized['face_found'] != 0)]

df_labels_recognized_cleaned.reset_index(drop=True, inplace=True)
```

```python
landmarks_df = gather_landmark_data(df_labels_recognized_cleaned, path_data, num_images=None)
```

```python
columns = ['normalized_left_eye', 'normalized_right_eye', 'normalized_eye_midpoint', 'normalized_nose', 
           'normalized_left_mouth', 'normalized_right_mouth']
labels = ['Left Eye', 'Right Eye', 'eye_midpoint', 'Nose', 'Left Mouth', 'Right Mouth']
landmarks_df.dropna(inplace=True)
plot_multiple_normalized_landmarks(landmarks_df, columns, labels)
```
